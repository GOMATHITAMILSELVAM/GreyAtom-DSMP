```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([7,9,11,13,15])
```

```python
def grad_descent(x,y):
    N = float(len(y))
    slope_current=0 
    intercept_current=0
    iterate=10000
    learning_rate=0.05
    for i in range(iterate):
        y_current = (slope_current**2 * x) + (intercept_current)
        cost =sum([data**2 for data in (y-y_current)]) / (2*N)
        slope_gradient = -(1/N) * sum(x * (y - y_current))
        intercept_gradient = -(1/N) * sum(y - y_current)
        slope_current = slope_current - (learning_rate * slope_gradient)
        intercept_current = intercept_current - (learning_rate * intercept_gradient)
        print("slope={}, intercept={} ,cost_func={} ,iterations={}".format(slope_current,intercept_current,cost,i))
        #return slope_current, intercept_current, cost
```

```python
grad_descent(x,y)
```

```python
#CORRELATION HEAT MAP
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),cmap="YlGnBu",annot=True,vmax=1,vmin=0)
```

```python
#plotting multiple BOXPLOTS
cols=df.columns
figs,axes=plt.subplots(6,2,figsize=(20,20))
for i in range(0,6):
    for j in range(0,2):
        col=cols[i*2+j]
        sns.boxplot(df[col],ax=axes[i,j],linewidth=2.5,orient='h',color='c')
        
```

```python
#COUNT PLOT
count=df['quality'].value_counts()
count
```

```python
sns.countplot(df['quality'],data=df)
```

```python
#HISTOGRAMS/DIST PLOTS
cols=df.columns
fig,axes=plt.subplots(6,2,figsize=(20,20))
for i in range(6):
    for j in range(2):
        c=cols[i*2+j]
        sns.distplot(df[c],ax=axes[i,j],kde=True,color='r')
```

```python
#STRIP PLOTS
cols=df.columns
fig,axes=plt.subplots(6,2,figsize=(20,20))
for i in range(6):
    for j in range(2):
        c=cols[i*2+j]
        sns.stripplot(df['quality'],df[c],ax=axes[i,j])
        
```

```python
#SWARM PLOTS
cols=df.columns
fig,axes=plt.subplots(6,2,figsize=(20,20))
for i in range(6):
    for j in range(2):
        c=cols[i*2+j]
        sns.swarmplot(df['quality'],df[c],ax=axes[i,j])
        
```

```python
#VIOLIN PLOTS
cols=df.columns
fig,axes=plt.subplots(6,2,figsize=(20,20))
for i in range(6):
    for j in range(2):
        c=cols[i*2+j]
        sns.violinplot(df['quality'],df[c],ax=axes[i,j],palette='rainbow',data=df)
```
